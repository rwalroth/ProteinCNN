```python
from tensorflow import keras
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
```

```python
from models import model_2d, model_1d, model_1d_rnn, model_1d_rnn_2
```

```python
from seq2mat import DataGenerator, seq_class, label_dict
```

```python
model_1d.load_weights('model_1d.model.best.hdf5')
model_2d.load_weights('model_2d.model.best.hdf5')
model_1d_rnn.load_weights('model_1d_rnn.model.best.hdf5')
model_1d_rnn_2.load_weights('model_1d_rnn_2.model.best.hdf5')
```

```python
train_val, test, y_train_val, y_test = train_test_split(seq_class.index, seq_class['label'], test_size=0.1, random_state=42)
```

```python
params_1d = {
        'batch_size': 4, 
        'dim': (1502, 31),
        'n_classes': 25, 
        'shuffle': False,
        'struct': '1d',
        'random_insert': False
    }

test_generator_1d = DataGenerator(test, seq_class['label'], **params_1d)
```

```python
class_dict = {val: key for key, val in label_dict.items()}
```

```python
preds = [class_dict[x] for x in model_1d.predict_generator(test_generator_1d, verbose=1, use_multiprocessing=True, workers=4).argmax(axis=-1)]
```

```python
params_2d = {
        'batch_size': 4, 
        'dim': (3072, 8, 6),
        'n_classes': 25, 
        'shuffle': False,
        'struct': '2d'
    }

test_generator_2d = DataGenerator(test, seq_class['label'], **params_2d)

preds = [class_dict[x] for x in model_2d.predict_generator(test_generator_2d, verbose=1, use_multiprocessing=True, workers=4).argmax(axis=-1)]

print(classification_report(seq_class['classification'][test], preds))
```

```python
print(classification_report(seq_class['classification'][test][test_generator_2d.indexes], preds))
```

```python
params_1d_rnn = {
        'batch_size': 4, 
        'dim': (1502, 31),
        'n_classes': 25, 
        'shuffle': True,
        'struct': '1d',
        'random_insert': False,
        'rnn': True
    }

test_generator_1d_rnn = DataGenerator(test, seq_class['label'], **params_1d_rnn)

preds = [class_dict[x] for x in model_1d_rnn.predict_generator(test_generator_1d_rnn, verbose=1, use_multiprocessing=True, workers=4).argmax(axis=-1)]

print(classification_report(seq_class['classification'][test][test_generator_1d_rnn.indexes], preds))
```
